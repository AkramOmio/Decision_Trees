# Train | Test Split

```python
X=pd.get_dummies(df.drop('species',axis=1),drop_first=True)
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_pred=model.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report,confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test,base_pred, labels=model.classes_)
Confusion=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
```

```python
Confusion.plot()
```

```python
print(classification_report(y_test,base_pred))
```

```python
model.feature_importances_
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance'])
```

```python
sns.boxplot(x='species',y='body_mass_g',data=df)
```

# Visualize the Tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8))
plot_tree(model);
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(model,filled=True,feature_names=X.columns);
```

```python
def report_model(model):
    model_preds=model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=150)
    plot_tree(model,filled=True,feature_names=X.columns);
```

# Max depth

```python
pruned_tree=DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

# Max Leaf Nodes

```python
pruned_tree=DecisionTreeClassifier(max_leaf_nodes=2)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

# Criterion

```python
entropy_tree=DecisionTreeClassifier(criterion='entropy')
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
